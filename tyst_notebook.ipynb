````python
# Setup Environment and Imports
from pathlib import Path
import typst
from IPython.display import SVG, display

# Define the temporary directory and file paths
temp_dir = Path.cwd()
typst_file = temp_dir / "hello.typ"
svg_file = temp_dir / "hello.svg"

# Define the content of the Typst file
file_content = """
#set page(width: auto, height: auto, margin: 0cm, fill: none)
#set text(size: 50pt)
$ sum_(k=1)^n k = (n(n+1)) / 2 $
"""

# Write the content to the Typst file
typst_file.write_text(file_content)

# Compile the Typst file to SVG format
typst.compile(typst_file, format="svg", output=str(svg_file))

# Display the generated SVG file
display(SVG(filename=str(svg_file)))
````
