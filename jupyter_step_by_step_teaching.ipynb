# Task 1)  $e^x$ and $x^2$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['lines.linewidth'] = 2
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
x = np.linspace(0, 5, 1001)
y = np.exp(x)
ax.plot(x, y, label=r"$e^x$")
ax.plot(x, x**2, label= r"$x^2$")
ax.legend(loc="lower right", fontsize=14);
```

**Question:**  
Which function grows faster: $e^x$ or $x^2$ ?  
*Note: Please plot both functions in the above cell for comparison*

- [x] $e^x$
- [ ] $x^2$
